```python
def calculate_angle(a,b,c):
    a = np.array(a)
    b = np.array(b)
    c = np.array(c)

    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians*180.0/np.pi)

    if angle > 180.0:
        angle = 360-angle
    return angle
```

```python
import cv2
import numpy as np
import mediapipe as mp

# Utility to calculate angle between three points
def calculate_angle(a, b, c):
    a = np.array(a)
    b = np.array(b)
    c = np.array(c)
    ba = a - b
    bc = c - b
    cosine_angle = np.dot(ba, bc) / (np.linalg.norm(ba) * np.linalg.norm(bc) + 1e-8)
    angle = np.degrees(np.arccos(np.clip(cosine_angle, -1.0, 1.0)))
    return angle

# Initialize video capture
cap = cv2.VideoCapture(0)

# Initialize Mediapipe Pose
mp_pose = mp.solutions.pose
mp_drawing = mp.solutions.drawing_utils

with mp_pose.Pose(min_detection_confidence=0.5,
                  min_tracking_confidence=0.5) as pose:
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            print("Failed to grab frame")
            break

        # Flip and convert to RGB for Mediapipe
        frame = cv2.flip(frame, 1)
        image_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image_rgb.flags.writeable = False

        # Perform pose detection
        results = pose.process(image_rgb)

        # Convert back to BGR and allow writing
        image_rgb.flags.writeable = True
        image_bgr = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2BGR)

        h, w, _ = image_bgr.shape
        if results.pose_landmarks:
            lm = results.pose_landmarks.landmark

            # Left side points
            ls = [lm[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x * w,
                  lm[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y * h]
            le = [lm[mp_pose.PoseLandmark.LEFT_ELBOW.value].x * w,
                  lm[mp_pose.PoseLandmark.LEFT_ELBOW.value].y * h]
            lw = [lm[mp_pose.PoseLandmark.LEFT_WRIST.value].x * w,
                  lm[mp_pose.PoseLandmark.LEFT_WRIST.value].y * h]
            lh = [lm[mp_pose.PoseLandmark.LEFT_HIP.value].x * w,
                  lm[mp_pose.PoseLandmark.LEFT_HIP.value].y * h]

            # Right side points
            rs = [lm[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x * w,
                  lm[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y * h]
            re = [lm[mp_pose.PoseLandmark.RIGHT_ELBOW.value].x * w,
                  lm[mp_pose.PoseLandmark.RIGHT_ELBOW.value].y * h]
            rw = [lm[mp_pose.PoseLandmark.RIGHT_WRIST.value].x * w,
                  lm[mp_pose.PoseLandmark.RIGHT_WRIST.value].y * h]
            rh = [lm[mp_pose.PoseLandmark.RIGHT_HIP.value].x * w,
                  lm[mp_pose.PoseLandmark.RIGHT_HIP.value].y * h]

            # Compute angles
            left_elbow_angle = calculate_angle(ls, le, lw)
            left_shoulder_angle = calculate_angle(le, ls, lh)
            right_elbow_angle = calculate_angle(rs, re, rw)
            right_shoulder_angle = calculate_angle(re, rs, rh)

            # Overlay angles on frame
            cv2.putText(image_bgr, f"L Elbow: {int(left_elbow_angle)}",
                        tuple(np.array(le, int)), cv2.FONT_HERSHEY_SIMPLEX,
                        0.5, (255, 255, 255), 2, cv2.LINE_AA)
            cv2.putText(image_bgr, f"L Shoulder: {int(left_shoulder_angle)}",
                        tuple(np.array(ls, int)), cv2.FONT_HERSHEY_SIMPLEX,
                        0.5, (255, 255, 255), 2, cv2.LINE_AA)
            cv2.putText(image_bgr, f"R Elbow: {int(right_elbow_angle)}",
                        tuple(np.array(re, int)), cv2.FONT_HERSHEY_SIMPLEX,
                        0.5, (255, 255, 255), 2, cv2.LINE_AA)
            cv2.putText(image_bgr, f"R Shoulder: {int(right_shoulder_angle)}",
                        tuple(np.array(rs, int)), cv2.FONT_HERSHEY_SIMPLEX,
                        0.5, (255, 255, 255), 2, cv2.LINE_AA)

        # Draw landmarks
        mp_drawing.draw_landmarks(image_bgr,
                                  results.pose_landmarks,
                                  mp_pose.POSE_CONNECTIONS)

        # Display the resulting frame
        cv2.imshow('Pose Angle Visualization', image_bgr)

        # Exit on 'q'
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import mediapipe as mp

# 3D Angle calculator
def calculate_angle_3d(a, b, c):
    """
    Calculates the angle at point b formed by points a–b–c in 3D.
    a, b, c: (x, y, z) sequences (e.g. landmark.x, .y, .z)
    """
    a = np.array(a)
    b = np.array(b)
    c = np.array(c)
    ba = a - b
    bc = c - b
    # dot product and norms in 3D
    cos_angle = np.dot(ba, bc) / (np.linalg.norm(ba) * np.linalg.norm(bc) + 1e-8)
    angle = np.degrees(np.arccos(np.clip(cos_angle, -1.0, 1.0)))
    return angle

# Initialize video + MediaPipe
cap = cv2.VideoCapture(0)
mp_pose = mp.solutions.pose
mp_drawing = mp.solutions.drawing_utils

with mp_pose.Pose(min_detection_confidence=0.5,
                  min_tracking_confidence=0.5) as pose:
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        frame = cv2.flip(frame, 1)  # mirror
        img_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        img_rgb.flags.writeable = False

        results = pose.process(img_rgb)

        img_rgb.flags.writeable = True
        image_bgr = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2BGR)
        h, w, _ = image_bgr.shape

        if results.pose_landmarks:
            lm = results.pose_landmarks.landmark

            # Grab full 3D landmarks (normalized)
            def landmark_3d(name):
                l = lm[mp_pose.PoseLandmark[name].value]
                return (l.x, l.y, l.z)

            # Example: left elbow angle in 3D
            ls = landmark_3d('LEFT_SHOULDER')
            le = landmark_3d('LEFT_ELBOW')
            lw = landmark_3d('LEFT_WRIST')
            left_elbow_angle_3d = calculate_angle_3d(ls, le, lw)

            # And left shoulder 3D angle
            lh = landmark_3d('LEFT_HIP')
            left_shoulder_angle_3d = calculate_angle_3d(le, ls, lh)

            # Overlay to screen (projected 2D positions for text)
            pts2d = {
                'LE': (int(lm[mp_pose.PoseLandmark.LEFT_ELBOW.value].x * w),
                       int(lm[mp_pose.PoseLandmark.LEFT_ELBOW.value].y * h)),
                'LS': (int(lm[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x * w),
                       int(lm[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y * h))
            }
            cv2.putText(image_bgr,
                        f"L3D Elbow: {int(left_elbow_angle_3d)}°",
                        pts2d['LE'], cv2.FONT_HERSHEY_SIMPLEX, 0.5,
                        (255,255,255), 2, cv2.LINE_AA)
            cv2.putText(image_bgr,
                        f"L3D Shoulder: {int(left_shoulder_angle_3d)}°",
                        pts2d['LS'], cv2.FONT_HERSHEY_SIMPLEX, 0.5,
                        (255,255,255), 2, cv2.LINE_AA)

        # Draw skeleton
        mp_drawing.draw_landmarks(image_bgr,
                                  results.pose_landmarks,
                                  mp_pose.POSE_CONNECTIONS)

        cv2.imshow("3D Pose Angles", image_bgr)
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```
